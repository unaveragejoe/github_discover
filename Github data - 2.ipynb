```python
def get_commit_data(commit_link_list):
    commit_dict = {}
    
    z = requests.get('https://api.github.com/rate_limit').json()
    remaining = z['resources']['core']['remaining']
        
    for k, v in commit_link_list:

        if remaining <= 0:
            time.sleep(3600)
            z = requests.get('https://api.github.com/rate_limit').json()
            remaining = z['resources']['core']['remaining']
        
        headers = {'User-Agent': ua.random}
  
        response = requests.get(v, headers)

        commits_json = json.loads(response.text)
        commit_dict[k] = commits_json
        
        z = requests.get('https://api.github.com/rate_limit').json()
        remaining = z['resources']['core']['remaining']
    
        timeDelay = random.randrange(2000, 4000)/1000
        time.sleep(timeDelay) 

        print(commits_json[:2])
        print("Scraped " + str(v)) 
        print("Waited " + str(timeDelay))
                            
    return commit_dict
```

```python
commit_dict_data = get_commit_data(commit_link_list)
```

```python
for k, v in commit_link_list:
    headers = {'User-Agent': ua.random}
    z = requests.get('https://api.github.com/rate_limit').json()
    remaining = z['resources']['core']['remaining']
    
    while remaining > 0:
        response = requests.get(v, headers)

        commits_json = json.loads(response.text)
        commit_dict[k] = commits_json

        timeDelay = random.randrange(2000, 4000)/1000
        time.sleep(timeDelay) 

        print("Scraped " + str(v)) 
        print("Waited " + str(timeDelay))
```

```python
def getCommits(list_of_repo_links):
    result = {}
    
    remaining = g.get_rate_limit().search.remaining
    
    for k,v in enumerate(list_of_repo_links):
          
        if remaining <= 0:
            time.sleep(600)
            remaining = g.get_rate_limit().search.remaining
            
        commits = g.search_commits(query='id:'+str(v[0]))
        for idx, commit in enumerate(commits):  
            result[v[0]] = commit.raw_data
            
        z = requests.get('https://api.github.com/rate_limit').json()
        remaining = g.get_rate_limit().search.remaining
        
        print(k)
    return result
```
